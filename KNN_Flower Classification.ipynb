```python
#gray image feature extraction
gray_image = []
for i in b:
    image = imread(i, as_gray=True)
    new_image = resize(image,(28, 28))
    features = np.reshape(new_image, new_image.shape[0]*new_image.shape[1])
    gray_image.append(features)
```

```python
gray_df = pd.DataFrame(gray_image)  
gray_df.head()
```

```python
gray_df.shape
```

```python
#rgb image feature extraction
rgb_image = []
for p in b:
    image = imread(p)
    new_image = resize(image,(28, 28))
    feature_matrix = np.zeros((new_image.shape[0], new_image.shape[1]))
    for i in range(0,new_image.shape[0]):
        for j in range(0,new_image.shape[1]):
            feature_matrix[i][j] = ((new_image[i,j,0] + new_image[i,j,1] + new_image[i,j,2])/3)
    features = np.reshape(feature_matrix, feature_matrix.shape[0]*feature_matrix.shape[1])
    rgb_image.append(features)
```

```python
rgb_df = pd.DataFrame(rgb_image)  
rgb_df.head()
```

```python
Folders = ["Daffodil","Snowdrop","LilyValley","Bluebell","Crocus","iris","Tigerlily","Tulip","Fritilary",
          "Sunflower","Daisy","ColtsFoot","Dandelion","Cowslip","Buttercup","Windflower","Pansy"]
```

```python
#Adding target Variable to extracted rgb data
a = 0
b = 79
for i in Folders:
    rgb_df.loc[a:b, "Target"] = i
    a = b+1
    b = b+80
```

```python
#Adding target Variable to ectracted gray data
a = 0
b = 79
for i in Folders:
    gray_df.loc[a:b, "Target"] = i
    a = b+1
    b = b+80
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score, accuracy_score, classification_report
```

```python
#KNN on rbg data
X = rgb_df.drop(['Target'], axis=1)
y = rgb_df['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.30, random_state = 15)
KNN = KNeighborsClassifier(n_neighbors=37)
KNN.fit(X_train,y_train)
KNN_pred = KNN.predict(X_test)
KNN_accuracy = accuracy_score(y_test, KNN_pred)
KNN_accuracy
```

```python
#KNN on gray data
X = gray_df.drop(["Target"],axis=1)
y = rgb_df["Target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.30, random_state = 15)
KNN = KNeighborsClassifier(n_neighbors=37)
KNN.fit(X_train,y_train)
KNN_pred = KNN.predict(X_test)
KNN_accuracy = accuracy_score(y_test, KNN_pred)
KNN_accuracy
```
